## Стек

```python
class stek():
    def __init__(self, l):# Инициализируем переменную 
        self.main_t = list(l)
    
    def ad(self, s):
        self.main_t.append(s) # Добавляем в конец
    
    def po(self):
        a = self.main_t[len(self.main_t)-1] #Сохраняем последний, чтобы вернуть
        del self.main_t[len(self.main_t)-1]
        return a
    
    def is_nan(self):# проверка по количеству элементов
        if len(self.main_t):
            return 'is not empty'
        else:
            return 'is empty'
        
    def le(self):# Считаем количество
        return len(self.main_t)
    
    def __str__(self):# Делаем красивый вывод
        return ', '.join(list(map(lambda x: str(x),self.main_t)))
```

```python
test1 = stek([1,2,3])
print('Проверка самого стека')
print(test1)
print('Добавления элемента')
test1.ad(4)
print(test1)
print('Удаление с возвращением')
print(test1.po())
print(test1)
print('Проверка на пустоту')
print(test1.is_nan())
print('Длина')
print(test1.le())
```

## Односвязный список

```python
class Node(): # Создаем класс, который будет содержать в себе значение и индекс
    
    def __init__(self, e, i = None):
        self.element = e
        self.next_e = i
        
        
class odnonap(): 
    
    def __init__(self, s = list()):
        self.sp1 = s # Тут странные присвоения, но они нужны для правильной работы 
        self.sp = list()
        for k, v in enumerate(self.sp1):
            if k != len(self.sp1)-1:
                self.sp.append(Node(v, k+1))
            else:
                self.sp.append(Node(v, None))
    
    def ad(self, item, number):
        
        
        if number != len(self.sp):
            self.sp.insert(number,Node(item, number)) # Вставляем элемент на нужное место
            for i in range(number, len(self.sp)-1): 
                self.sp[i].next_e = self.sp[i].next_e + 1 # Изменяем адреса
        else:
            self.sp[len(self.sp)-1].next_e = number # Учитываем вставку в конец
            self.sp.insert(number,Node(item, None))
            
            
    def pr(self): # печать списка, для проверки результатов 
        print(list(map(lambda x: [x.element, x.next_e], self.sp)))
    
    def le(self): # Длина
        return len(self.sp)
    
    def is_empty(self): # Пустота 
        return self.sp == list()
    
    def dell(self, number): # Удаление 
        a = self.sp.pop(number).element # Удаляем и возвращаем 
        
        if number != len(self.sp):
            for i in range(number, len(self.sp)-1):
                self.sp[i].next_e = self.sp[i].next_e - 1 # Попрпавляем адреса 
        else:
            self.sp[len(self.sp)-1].next_e = None # Учитываем последний 
        
        return a
```

```python
a = odnonap(["a","b","c","d"])
print('Весь список')
a.pr()
print('Добавляем')
a.ad('l',4)
a.pr()
print('Удаляем с возвращением')
print(a.dell(3))
a.pr()
print('Проверка на пустоту')
print(a.is_empty())
print('Длина')
print(a.le())
```

## Двунаправленный список

```python
class Node2(): # Создаем класс, который будет содержать в себе значение и индексы
    
    def __init__(self, s, i, j):
        self.element = s
        self.next = i  # индекс следующего
        self.back = j  # индекс предыдущего 
    
class tworoad():
    
    def __init__(self, s):
        self.sp1 = s
        self.sp = list() # тут странные преобразования, но надо чтобы правильно работало
        for k, v in enumerate(self.sp1) :
            if (k != len(self.sp1)-1) and k != 0:
                self.sp.append(Node2(v, k+1, k-1))
            if k == len(self.sp1)-1:
                self.sp.append(Node2(v, None, k-1))
            if k == 0:
                self.sp.append(Node2(v, k+1, None))
    def pr(self): # Печать для проверки
        print(list(map(lambda x: [x.back,x.element, x.next], self.sp)))
        
    def le(self): # Длина
        return len(self.sp)
    
    def is_empty(self): # Проверка на пустоту
        return self.sp == list()
    
    def ad(self, item, number): # Добавление

        if number != len(self.sp) and number != 0: # добавление в середину 
            self.sp.insert(number, Node2(item, number+1 , number-1))
            for i in range(number+1, len(self.sp)-1):# фикс индексов
                self.sp[i].next = self.sp[i].next + 1
                self.sp[i].back = self.sp[i].back + 1
            self.sp[len(self.sp)-1].back += 1
        if number == len(self.sp):#  добавление в конец
            self.sp[len(self.sp)-1].next = number # Фиксим индекс предпоследнего 
            self.sp.insert(number,Node2(item, None, number-1))# добавляем в конец элемент
        if number == 0:# Добавление в конец 
            self.sp[0].back = number # фиксим адреса 
            self.sp[0].next = 2
            self.sp.insert(number,Node2(item, number + 1, None)) # добавляем элемент 
            for i in range(number+2, len(self.sp)-1):# опять фиксим индексы 
                self.sp[i].next = self.sp[i].next + 1
                self.sp[i].back = self.sp[i].back + 1
            self.sp[len(self.sp)-1].back += 1
            
    def dell(self, number):# Удаляем, тут в принципе всё то же самое, как при добавлении, только вместо плюсов минусы
        a = self.sp.pop(number).element
        if number != len(self.sp) and number != 0: # из середины
            for i in range(number+1, len(self.sp)-1):
                self.sp[i].next = self.sp[i].next - 1
                self.sp[i].back = self.sp[i].back - 1
            self.sp[len(self.sp)-1].back -= 1
        if number == len(self.sp): # в конце 
            self.sp[len(self.sp)-1].next = None
        if number == 0: # в начало 
            for i in range(number+1, len(self.sp)-1):
                self.sp[i].next = self.sp[i].next - 1
                self.sp[i].back = self.sp[i].back - 1
            self.sp[len(self.sp)-1].back -= 1
            self.sp[0].back = None
            self.sp[0].next = 1
        return a
```

```python
a = tworoad(['q','w','e','r'])
print('Весь список')
a.pr()
print('Добавляем')
a.ad('l',4)
a.pr()
print('Удаляем с возвращением')
print(a.dell(3))
a.pr()
print('Проверка на пустоту')
print(a.is_empty())
print('Длина')
print(a.le())
```
